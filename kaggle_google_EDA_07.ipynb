# Columns Break Down
1. App: Name of app and uses.
2. Category: For which purpose this app is design (eg: art,economics,entertainment).
3. Rating: The pupolarity of app.
4. Reviews: The people reviews for app.
5. Size: The size of app.
6. Install: How many people install it.
7. Type: Is it free or not.
8. Price: Price of app.
9. Content Rating: Which type of people can use it(eg:adult,kids etc).
10. Genres: Description of app or purpose.
11. Last_upadate: The date of last update of the app.
12. Current_ver: The current version of the app.
13. Andriod_ver: The cureent version of the Andriod_ver.
