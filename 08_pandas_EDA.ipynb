```python
# we need to convert kb into Mbs and then remove M from all datapoints
# ...existing code...
def size_to_mb(size):
    if size[-1] == 'M':
        return float(size[:-1])
    elif size[-1] == 'k':
        return float(size[:-1]) / 1024
    else:
        return np.nan  # or handle other cases as needed

df['Size_MB'] = df['Size_MB'].replace('Varies with device', np.nan)
df['Size_MB'] = df['Size'].dropna().apply(size_to_mb)
# ...existing code...

df.head(20)
```
